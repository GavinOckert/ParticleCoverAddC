```python
from src.readers.reader import * 
from src.coverers.data_structs import * 
from src.coverers.line import * 
from src.coverers.wedgecover import *
from src.testers.test import *
import matplotlib.pyplot as plt 
import time 
import ast
```

```python
filepath = "data/wedgeData_v3_128.txt"
f = open('data/v3_patches.txt')
filedata = readFile(filepath, stop=128, performance=False)
env, points = filedata[0]
ds = DataSet(env)
ds.importData(points)
ds.add()
apexZ0 = [0]
colors = ['orange', 'b', 'm']
plt.figure(figsize = (10, 8))

for i in range(len(apexZ0)):
    cov = wedgeCover(env, ds)
    cov.makePatches_Projective_center(apexZ0 = apexZ0[i])
    fitting_lines = []
    lGen = LineGenerator(env, 10)
    lines_to_plot = []
    fitting_lines = fitting_lines + lGen.generateEvenGrid(100)
    for j, patch in enumerate(cov.patches):
        top = env.radii[-1]
        end = patch.end_layer
        left = min(patch.superpoints[end-1].z_values)
        right = max(patch.superpoints[end-1].z_values)
        min_lambdaZ = (left - apexZ0[i])/env.radii[end-1]
        max_lambdaZ = (right - apexZ0[i])/env.radii[end-1]
        if j ==0:
            label = r"$apexZ_0$ = " + f"{apexZ0[i]}"
        else:
            label = '_'
        plt.plot([apexZ0[i],min_lambdaZ*top, max_lambdaZ*top, apexZ0[i]], [0, top, top, 0], color = colors[i], label = label)
        plt.fill_between([apexZ0[i],min_lambdaZ*top, max_lambdaZ*top, apexZ0[i]], [0, top, top, 0], alpha = 0.2, color =colors[i])
        for line_index, line in enumerate(fitting_lines):
            if patch.contains(line) == True:
                lines_to_plot.append(line_index)
    for l, line in enumerate(fitting_lines):
        if l in lines_to_plot:
            #line.plot('g')
            pass
        else:
            line.plot('r')
plt.plot([0],color = 'r', label = 'Lines Not Accepted')
plt.legend()
ds.plot(True, True)
```
